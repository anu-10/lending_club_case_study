# Import dependencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load Dataset

```python
df = pd.read_csv('loan.csv', engine='python')
df.head()
```

```python
df.shape
```

# Data Cleaning

<b>Removing Empty columns</b>

```python
res = []
for col in df.columns:
    if df[col].isnull().sum() == len(df):
        res.append(col)
len(res)
```

```python
for col in res:
    df.drop(col, axis = 1, inplace=True)
```

```python
df.isnull().sum()[df.isnull().sum() > 0]
```

#### Dropping columns with large null values or single value

```python
remove_list = ['mths_since_last_delinq', 'mths_since_last_record', 'next_pymnt_d']
print('Dropped Columns ' +str(remove_list))
for i in df.columns:
    if df[i].nunique() == 1:
        print(i)
        remove_list.append(i)
df.drop(remove_list, axis = 1, inplace = True)
```

```python
df.isnull().sum()
```

```python
df['loan_status'].value_counts()
```

<b>Current would not help in identifying loan defaults so dropping it</b>

```python
df = df[df['loan_status'] != 'Current']
```

```python
sorted(df.columns)
```

#### Selecting useful columns

```python
cols = ['annual_inc','dti','emp_length','funded_amnt','funded_amnt_inv','grade',
        'home_ownership','inq_last_6mths','installment','int_rate','issue_d','loan_amnt',
        'loan_status','pub_rec_bankruptcies','purpose','sub_grade','term','verification_status']
```

```python
df = df[cols]
```

```python
df.dtypes
```

<b>Converting objects to numerical values for easier analysis</b>

```python
df.term.value_counts()
```

```python
df.term = df.term.apply(lambda x: int(x.replace('months', '')))
```

```python
df.int_rate = df.int_rate.apply(lambda x: float(x.replace('%', '')))
```

### Imputing Values

#### Employee Length
NA could signify that the person is self-employeed or doing business

```python
df.emp_length.isna().sum()
```

```python
df[df.emp_length.isna()]['annual_inc'].describe()
```

```python
df.emp_length.mode()
```

```python
val = df.emp_length.mode()[0]
df.emp_length = df.emp_length.fillna(val)
```

```python
emp_length_mapping: dict = {
    '< 1 year': 0, '1 year': 1, '2 years': 2, '3 years': 3,
    '4 years': 4, '5 years': 5, '6 years': 6, '7 years': 7,
    '8 years': 8, '9 years': 9, '10+ years': 10
}
```

```python
df['emp_length'] = df['emp_length'].map(emp_length_mapping)
```

There are 3 NONE values which can be grouped up with OTHER

```python
df.home_ownership.value_counts()
```

```python
df['home_ownership'] = df['home_ownership'].apply(lambda x : x.replace('NONE', 'OTHER'))
```

```python
df.home_ownership.value_counts()
```

```python
df.verification_status.value_counts()
```

Source Verified also comes under Verified so grouping them

```python
df['verification_status'] = df['verification_status'].apply(lambda x : x.replace('Source Verified', 'Verified'))
```

```python
df.verification_status.value_counts()
```

```python
df.pub_rec_bankruptcies.isna().sum()
```

pub_rec_bankruptcies cannot be imputed as NA does not have suitable replacement

```python
df = df[df.pub_rec_bankruptcies.notna()]
```

```python
df.isna().sum()
```

Adding issue year and issue month from issue_d column for better analysis

```python
df['issue_y'] = df.issue_d.apply(lambda x : 2000+int(x.split('-')[1]))
```

```python
df['issue_m'] = df.issue_d.apply(lambda x : x.split('-')[0])
```

```python
# function to assign bin value to column
def bin_data(column, thresholds, labels) -> str:
    for i, threshold in enumerate(thresholds):
        if column < threshold:
            return labels[i]
    return labels[-1]
```

```python
df.int_rate.plot.hist()
plt.title('Histogram of Interest Rate')
plt.xlabel('Interest Rate')
plt.show()
```

```python
int_rate_bins = [5, 10, 15, 20]
int_rate_labels = ['0%-5%', '5%-10%','10%-15%','15%-20%','20%-25%']
df['int_rate_bin'] = df['int_rate'].apply(bin_data, args=((int_rate_bins, int_rate_labels)))
```

```python
df.loan_amnt.plot.hist()
plt.title('Histogram of Loan Amount')
plt.xlabel('Loan Amount')
plt.show()
```

```python
loan_amnt_bins = [5000, 10000, 20000, 30000]
loan_amnt_labels = ['0 - 5K', '5K - 10K', '10K - 20K', '20K - 30K', '30K - above']
df['loan_amnt_bin'] = df['loan_amnt'].apply(bin_data, args=((loan_amnt_bins, loan_amnt_labels)))
```

```python
df.funded_amnt.plot.hist()
plt.title('Histogram of Funded Amount')
plt.xlabel('Funded Amount')
plt.show()
```

```python
funded_amnt_bins = [5000, 10000, 20000, 30000]
funded_amnt_labels = ['0 - 5K', '5K - 10K', '10K - 20K', '20K - 30K', '30K - above']
df['funded_amnt_bin'] = df['funded_amnt'].apply(bin_data, args=((funded_amnt_bins, funded_amnt_labels)))
```

```python
df.dti.plot.hist()
plt.title('Histogram of Debt to Interest')
plt.xlabel('Debt to Interest')
plt.show()
```

```python
dti_bins = [5, 10, 15, 20]
dti_labels = ['0%-5%', '5%-10%','10%-15%','15%-20%','20%+']
df['dti_bin'] = df['dti'].apply(bin_data, args=((dti_bins, dti_labels)))
```

```python
df.boxplot('annual_inc')
plt.title('Boxplot of Annual Income')
plt.show()
```

We can see that there are many outliers in this boxplot. So removing them would help in the analysis.

```python
q1 = df.annual_inc.quantile(0.25)
q3 = df.annual_inc.quantile(0.75)
iqr = q3-q1
ub = q3 + iqr * 1.5
ub
```

```python
df = df[df['annual_inc'] < ub]
```

```python
annual_inc_bins = [30000, 60000, 90000 , 120000]
annual_inc_labels = ['0 - 30K', '30K - 60K', '60K - 120K', '120K above']
df.loc[:, 'annual_inc_bin'] = df['annual_inc'].apply(bin_data, args=((annual_inc_bins, annual_inc_labels)))
```

```python
def barplot(vis_df, title, xtitle, ytitle, figsize=(8,6)):
    plt.figure(figsize=figsize)
    sns.barplot(vis_df)
    
    plt.title(title, fontsize=14)
    plt.xlabel(xtitle, fontsize=12)
    plt.ylabel(ytitle, fontsize=12)
    plt.show()
```

## Summary
- Loan applicants in Grades B, C, and D account for the majority of "charged off" loans.

- Loan applicants from Sub Grades B3, B4, and B5 are more likely to charge off.

- Loan applicants who seek for a 60-month term are more likely to default than those who take out a 36-month loan.

- The majority of loan applicants have 10 or more years of experience, and they are the most likely to default.

- The number of loan applications climbed steadily between 2007 and 2011, predicting a healthy trend in the next years.

- December is the most popular month for taking out loans, presumably because of the holiday season.


- Loan applicants who reside in rented or mortgaged homes are more likely to default.

## Inferences
- Term Length: Given that applicants who choose for 60-month loans are more likely to default, the organisation should consider assessing the risk associated with longer-term loans and either limiting the maximum term or modifying interest rates accordingly.

- Risk Assessment for Grades B, C, and D: Because loan applicants from Grades B, C, and D account for the majority of "Charged Off" loans, the company should consider establishing stronger risk assessment and underwriting criteria for applicants in these grades.

- Pay close attention to applicants with Subgrades B3, B4, and B5, as they are more prone to charge off. Additional risk reduction strategies, such as granting lesser loan amounts, should be considered.

- Loan applicants with 10 or more years of experience have a higher likelihood of defaulting. This shows that experience alone may not be an accurate predictor of creditworthiness. The organisation should adopt a more complete credit rating system that considers other risk-related indicators.

- Debt Consolidation Risk: Because debt consolidation is the category with the most loans and the highest default rates, the company should carefully analyse applicants for debt consolidation loans and possibly change interest rates or provide financial counselling services.

- Positive expansion Trend: From 2007 to 2011, the number of loan applications increased steadily, indicating market expansion. The organisation may capitalise on this trend by maintaining a competitive advantage in the sector while implementing strong risk management techniques.

- Housing Status and Default Risk: Applicants who live in rented or mortgaged properties are more likely to default. This information can be used during the underwriting process to determine housing stability and its impact on repayment ability.

- Verification Process: Verified loan applicants default more frequently than those who are not verified. The organisation should review its verification process to ensure that it effectively assesses applicant creditworthiness, and make any necessary modifications or adjustments.
